### Design different node faces for ancestors and extants

```python
###############################################################################
# Date: 5/12
# Author: Sebastian Porras   
# This is the basic code that could be wrapped into an API call to the 
# bnkit server and generate some output for the user. I can package it
# up into a function and also have the option to maybe do inline or 
# an interactive interface depending on how interested the user is. 
################################################################################

def request_joint_tree(tree_file: "nwk", interactive: bool):
    """
    Creates a visualisation of the joint tree to allow easy visualisation
    currently needs a nwk file with ancestors, plan is to eventually
    input the aln and nwk and get back the reconstruction from the API 
    
    Parameters:
    
    tree_file (nwk): currently needs a nwk file with ancestor.
    
    interactive(bool): when True, initialises the ETE3 GUI, when False
                       simply render an inline figure. 
                       
    returns: None
    
    """
    
    
    t = Tree(tree_file, format=1)

    for node in t.traverse("postorder"):
        if not node.is_leaf():
            node.add_features(is_ancestor = True)
        else:
            node.add_features(is_ancestor = False)

    #for node in t.traverse("postorder"):
     #   print(node.name, node.is_ancestor)
      #  print()


    anc_nstyle = NodeStyle()
    anc_nstyle["fgcolor"] = "LightSteelBlue"
    anc_nstyle["size"] = 20
    anc_nstyle["vt_line_width"] = 5
    anc_nstyle["hz_line_width"] = 5


    leaf_nstyle = NodeStyle()
    leaf_nstyle["fgcolor"] = "ForestGreen"
    leaf_nstyle["size"] = 15
    leaf_nstyle["vt_line_width"] = 5
    leaf_nstyle["hz_line_width"] = 5


    for node in t.traverse("postorder"):
        if node.is_leaf():
            node.set_style(leaf_nstyle)
        else:
            node.set_style(anc_nstyle)
            name = TextFace(node.name,bold=True)
            name.margin_top = 10
            name.margin_right = 0
            name.margin_left = 15
            name.margin_bottom = 10
            node.add_face(name, column=0, position="branch-top")


    ts = TreeStyle()
    ts.show_branch_support = True
    ts.branch_vertical_margin = 10
    ts.show_leaf_name = True
    
    if interactive:
        return t.show()
    else:
        return t.render("%%inline", tree_style=ts)

request_joint_tree("_ancestors.nwk", False)
#request_joint_tree("_ancestors.nwk", True)
```
